```python
import numpy as np
import pandas as pd

# ── CLEAN DATA ─────────────────────────────────────────────────────────────
before = stock_prices.shape[0]
stock_prices_clean = stock_prices.dropna(axis=0)
after  = stock_prices_clean.shape[0]
print(f"Stocks before : {before} | Removed : {before - after} | After : {after}")

prices    = stock_prices_clean.values.astype(float)
n_stocks, n_time = prices.shape

# ── CONFIG ─────────────────────────────────────────────────────────────────
capital           = 100_000
unit_size         = 20_000
leverage          = 5
position_size     = unit_size * leverage       # ₹1,00,000 actual exposure

HARD_STOP_AMT     = position_size * 0.02       # 2% of ₹1L = ₹2,000
TRAIL_GAP         = 1_000                      # trailing stop pullback in ₹
TRAIL_ACTIVATE    = 500                        # trail activates earlier — ₹500 profit
BREAKEVEN_TRIGGER = 1_000                      # breakeven stop after ₹1k profit
DAILY_LOSS_LIMIT  = 5_000
VOLATILITY_CAP    = 3.0                        # spike filter %

# ── ENTRY PARAMETERS (Scenario A — early entry) ───────────────────────────
BAR1_MIN          = 1.0    # bar 1 must be strong (>1%) — quality filter
BAR2_MIN          = 0.0    # bar 2 just needs to agree direction (>0%)
# Logic: strong first bar + any confirmation = enter early into the move

EOD_BUFFER        = 15     # no entry/exit eval in last 15 bars
OPEN_BUFFER       = 30     # no entry in first 30 bars
MIN_HOLD_BARS     = 3      # minimum bars before any exit
COOLDOWN_BARS     = 10     # bars before re-entry on same stock

max_units         = 5
free_units        = max_units
positions         = []
closed_pnl        = []
trade_log         = []
daily_loss        = 0.0
trading_halted    = False
capital_available = capital
cooldown          = {}
recent_results    = []
stock_stats       = {}

# ── COST MODEL ────────────────────────────────────────────────────────────
def compute_costs(pos_val):
    buy_val = sell_val = pos_val
    brokerage    = 40
    stt          = 0.00025   * sell_val
    exchange_fee = 0.0000345 * (buy_val + sell_val)
    sebi         = 0.000001  * (buy_val + sell_val)
    stamp        = 0.00003   * buy_val
    gst          = 0.18 * (brokerage + exchange_fee + sebi)
    slippage     = 0.0005    * (buy_val + sell_val)
    return round(brokerage + stt + exchange_fee + sebi + stamp + gst + slippage, 2)

COST_PER_TRADE = compute_costs(position_size)
print(f"Hard stop      : ₹{HARD_STOP_AMT:,.0f}")
print(f"Trail gap      : ₹{TRAIL_GAP:,.0f} (activates at ₹{TRAIL_ACTIVATE:,.0f} profit)")
print(f"Entry signal   : bar1 > {BAR1_MIN}%,  bar2 > {BAR2_MIN}% (direction only)")
print(f"Exit           : trailing stop + hard stop only (exhaustion removed)")
print(f"Cost per trade : ₹{COST_PER_TRADE:,.2f}")

# ── HELPERS ───────────────────────────────────────────────────────────────
def safe_pct(curr, prev):
    with np.errstate(divide="ignore", invalid="ignore"):
        return np.where(
            (prev == 0) | np.isnan(prev) | np.isnan(curr),
            0.0,
            (curr - prev) / np.abs(prev) * 100
        )

def is_valid_price(p):
    return float(p) > 0 and not np.isnan(float(p))

def is_spike(p1, p2):
    return abs(p1) > VOLATILITY_CAP or abs(p2) > VOLATILITY_CAP

def calc_raw_pnl(trade_type, entry, curr):
    if not is_valid_price(entry) or not is_valid_price(curr):
        return 0.0
    return ((curr - entry) / entry * position_size if trade_type == "buy"
            else (entry - curr) / entry * position_size)

def get_entry_threshold_bar1():
    """Raise bar1 threshold after 3 consecutive losses."""
    if len(recent_results) >= 3 and not any(recent_results[-3:]):
        return BAR1_MIN + 0.5   # tighten to 1.5% after bad run
    return BAR1_MIN

def close_trade(pos, net_pnl, raw_pnl, curr_price, exit_reason, t):
    global free_units, capital_available, daily_loss, trading_halted

    recent_results.append(net_pnl > 0)
    closed_pnl.append(net_pnl)
    capital_available += unit_size
    free_units += 1
    cooldown[pos["stock"]] = t + COOLDOWN_BARS

    s = pos["stock"]
    if s not in stock_stats:
        stock_stats[s] = {"trades": 0, "pnl": 0.0, "wins": 0}
    stock_stats[s]["trades"] += 1
    stock_stats[s]["pnl"]    += net_pnl
    stock_stats[s]["wins"]   += int(net_pnl > 0)

    if net_pnl < 0:
        daily_loss += abs(net_pnl)
        if daily_loss >= DAILY_LOSS_LIMIT:
            trading_halted = True

    trade_log.append({
        "stock"        : int(s),
        "type"         : pos["type"],
        "entry_time"   : int(pos["entry_time"]),
        "exit_time"    : int(t),
        "bars_held"    : int(t - pos["entry_time"]),
        "entry_price"  : round(float(pos["entry_price"]), 2),
        "exit_price"   : round(float(curr_price), 2),
        "peak_pnl"     : round(float(pos["peak_pnl"]), 2),
        "raw_pnl"      : round(float(raw_pnl), 2),
        "costs"        : COST_PER_TRADE,
        "net_pnl"      : round(float(net_pnl), 2),
        "exit_reason"  : exit_reason,
        "daily_loss"   : round(daily_loss, 2),
        "breakeven_on" : pos.get("breakeven_active", False),
    })

# ── MAIN LOOP ─────────────────────────────────────────────────────────────
for t in range(31, n_time - 1):

    if trading_halted and len(positions) == 0:
        break

    pct_now  = safe_pct(prices[:, t],   prices[:, t-1])
    pct_prev = safe_pct(prices[:, t-1], prices[:, t-2])

    # ── EXIT ──────────────────────────────────────────────────────────────
    still_open = []

    for pos in positions:
        s          = pos["stock"]
        curr_price = prices[s, t]
        trade_type = pos["type"]
        bars_held  = t - pos["entry_time"]

        raw_pnl = calc_raw_pnl(trade_type, pos["entry_price"], curr_price)
        net_pnl = raw_pnl - COST_PER_TRADE

        # Update peak
        pos["peak_pnl"] = max(pos["peak_pnl"], raw_pnl)

        # Breakeven stop — once profit hits trigger, floor becomes entry price
        if raw_pnl >= BREAKEVEN_TRIGGER and not pos.get("breakeven_active"):
            pos["breakeven_active"] = True

        effective_hard_stop = (
            -COST_PER_TRADE      # breakeven = lose only the trading costs
            if pos.get("breakeven_active")
            else -HARD_STOP_AMT
        )

        trail_level = pos["peak_pnl"] - TRAIL_GAP

        # ── Exit conditions — exhaustion REMOVED ──────────────────────────
        is_force_eod  = (t >= n_time - 1 - EOD_BUFFER)
        is_hard_stop  = (net_pnl <= effective_hard_stop)
        is_trail_stop = (
            (raw_pnl < trail_level) and
            (pos["peak_pnl"] >= TRAIL_ACTIVATE) and
            (bars_held >= MIN_HOLD_BARS)        # minimum hold respected
        )

        if is_force_eod:
            close_trade(pos, net_pnl, raw_pnl, curr_price, "force_eod", t)
        elif is_hard_stop:
            close_trade(pos, net_pnl, raw_pnl, curr_price, "hard_stop", t)
        elif is_trail_stop:
            close_trade(pos, net_pnl, raw_pnl, curr_price, "trail_stop", t)
        else:
            still_open.append(pos)

    positions = still_open

    # ── ENTRY ─────────────────────────────────────────────────────────────
    in_open_window   = t < OPEN_BUFFER
    near_eod         = t >= (n_time - 1 - EOD_BUFFER)
    insufficient_cap = capital_available < unit_size

    if trading_halted or free_units == 0 or in_open_window or near_eod or insufficient_cap:
        continue

    bar1_threshold = get_entry_threshold_bar1()
    open_stocks    = {p["stock"] for p in positions}
    open_buys      = sum(1 for p in positions if p["type"] == "buy")
    open_sells     = sum(1 for p in positions if p["type"] == "sell")

    candidates = sorted(range(n_stocks),
                        key=lambda s: abs(pct_now[s]), reverse=True)

    for s in candidates:
        if free_units == 0 or capital_available < unit_size:
            break
        if s in open_stocks:
            continue
        if not is_valid_price(prices[s, t]):
            continue
        if cooldown.get(s, 0) > t:
            continue

        # p1 = previous bar (the strong signal bar)
        # p2 = current bar  (confirmation — direction only)
        p1 = pct_prev[s]
        p2 = pct_now[s]

        if p1 == 0.0 or p2 == 0.0:
            continue
        if is_spike(p1, p2):
            continue

        # BUY — strong bar 1, bar 2 confirms direction
        if p1 > bar1_threshold and p2 > BAR2_MIN:
            if open_buys >= 3:
                continue
            positions.append({
                "stock"            : s,
                "type"             : "buy",
                "entry_price"      : prices[s, t],
                "entry_time"       : t,
                "peak_pnl"         : 0.0,
                "breakeven_active" : False,
            })
            open_stocks.add(s)
            open_buys += 1
            capital_available -= unit_size
            free_units -= 1

        # SELL — strong bar 1 (negative), bar 2 confirms direction
        elif p1 < -bar1_threshold and p2 < -BAR2_MIN:
            if open_sells >= 3:
                continue
            positions.append({
                "stock"            : s,
                "type"             : "sell",
                "entry_price"      : prices[s, t],
                "entry_time"       : t,
                "peak_pnl"         : 0.0,
                "breakeven_active" : False,
            })
            open_stocks.add(s)
            open_sells += 1
            capital_available -= unit_size
            free_units -= 1

# ── RESULTS ───────────────────────────────────────────────────────────────
valid_pnl   = [p for p in closed_pnl if not np.isnan(p)]
total_pnl   = round(np.sum(valid_pnl), 2)
num_trades  = len(trade_log)
avg_pnl     = round(np.mean(valid_pnl), 2) if valid_pnl else 0
wins        = sum(1 for p in valid_pnl if p > 0)
losses      = sum(1 for p in valid_pnl if p <= 0)
total_costs = round(COST_PER_TRADE * num_trades, 2)
best_trade  = round(max(valid_pnl), 2) if valid_pnl else 0
worst_trade = round(min(valid_pnl), 2) if valid_pnl else 0
avg_bars    = round(np.mean([t["bars_held"] for t in trade_log]), 1) if trade_log else 0
be_count    = sum(1 for t in trade_log if t["breakeven_on"])

exit_counts = {}
for tr in trade_log:
    exit_counts[tr["exit_reason"]] = exit_counts.get(tr["exit_reason"], 0) + 1

print("\n" + "=" * 54)
print(f"  Total Net PnL      : ₹{total_pnl:>10,.2f}")
print(f"  Total Costs        : ₹{total_costs:>10,.2f}")
print(f"  Capital Remaining  : ₹{capital_available:>10,.2f}")
print(f"  Daily Loss         : ₹{daily_loss:>10,.2f}  "
      f"{'⚠ HALTED' if trading_halted else '✓ OK'}")
print("-" * 54)
print(f"  Trades             : {num_trades}"
      + (f"  ({wins}W / {losses}L)  WR: {wins/num_trades*100:.1f}%"
         if num_trades else ""))
print(f"  Avg PnL / Trade    : ₹{avg_pnl:>10,.2f}")
print(f"  Best Trade         : ₹{best_trade:>10,.2f}")
print(f"  Worst Trade        : ₹{worst_trade:>10,.2f}")
print(f"  Avg Bars Held      : {avg_bars} bars")
print(f"  Breakeven used     : {be_count} trades")
print("-" * 54)
for reason, count in sorted(exit_counts.items(), key=lambda x: -x[1]):
    pct = count / num_trades * 100 if num_trades else 0
    print(f"  Exit {reason:<16}: {count:>3}  ({pct:.0f}%)")
print("=" * 54)

# Per-stock breakdown
if stock_stats:
    df_stocks = pd.DataFrame([
        {"stock": s, "trades": v["trades"],
         "wins": v["wins"], "net_pnl": round(v["pnl"], 2)}
        for s, v in stock_stats.items()
    ]).sort_values("net_pnl", ascending=False)
    print("\nTop 5 stocks:")
    print(df_stocks.head(5).to_string(index=False))
    print("\nBottom 5 stocks:")
    print(df_stocks.tail(5).to_string(index=False))

print("\nFull Trade Log:")
df = pd.DataFrame(trade_log)
if not df.empty:
    print(df.to_string(index=False))
```

```python
import numpy as np
import pandas as pd

# ── CLEAN DATA ─────────────────────────────────────────────────────────────
stock_prices_clean = stock_prices.dropna(axis=0)
prices    = stock_prices_clean.values.astype(float)
n_stocks, n_time = prices.shape

# ── CONFIG ─────────────────────────────────────────────────────────────────
capital           = 100_000
unit_size         = 20_000
leverage          = 5
position_size     = unit_size * leverage       # ₹1,00,000 exposure

OPENING_BARS      = 30                         # first 30 bars = opening range
EOD_BUFFER        = 15                         # force close last 15 bars
HARD_STOP_PCT     = 0.015                      # 1.5% of entry price
TRAIL_GAP_PCT     = 0.008                      # 0.8% trailing stop gap
TRAIL_ACTIVATE_PCT= 0.010                      # trail activates after 1% profit
BREAKEVEN_PCT     = 0.015                      # breakeven after 1.5% profit
DAILY_LOSS_LIMIT  = 5_000
MIN_PRICE         = 10.0                       # filter penny stocks
MIN_RANGE_PCT     = 0.005                      # opening range must be >0.5% wide
MAX_RANGE_PCT     = 0.05                       # skip if opening range >5% (too volatile)
COOLDOWN_BARS     = 15
MAX_BUYS          = 3                          # max simultaneous buys
MAX_SELLS         = 3                          # max simultaneous sells

max_units         = 5
free_units        = max_units
positions         = []
closed_pnl        = []
trade_log         = []
daily_loss        = 0.0
trading_halted    = False
capital_available = capital
cooldown          = {}
stock_stats       = {}

# ── COST MODEL ────────────────────────────────────────────────────────────
def compute_costs(pos_val):
    buy_val = sell_val = pos_val
    brokerage    = 40
    stt          = 0.00025   * sell_val
    exchange_fee = 0.0000345 * (buy_val + sell_val)
    sebi         = 0.000001  * (buy_val + sell_val)
    stamp        = 0.00003   * buy_val
    gst          = 0.18 * (brokerage + exchange_fee + sebi)
    slippage     = 0.0005    * (buy_val + sell_val)
    return round(brokerage + stt + exchange_fee + sebi + stamp + gst + slippage, 2)

COST_PER_TRADE = compute_costs(position_size)

# ── COMPUTE OPENING RANGE FOR EVERY STOCK ─────────────────────────────────
# Opening range = high and low of first 30 bars
or_high = prices[:, :OPENING_BARS].max(axis=1)   # highest price in first 30 bars
or_low  = prices[:, :OPENING_BARS].min(axis=1)   # lowest price in first 30 bars
or_mid  = (or_high + or_low) / 2
or_range_pct = (or_high - or_low) / or_mid       # range as % of midpoint

# ── STOCK ELIGIBILITY FILTER ──────────────────────────────────────────────
# A stock is eligible if:
# 1. Opening price is above minimum (not a penny stock)
# 2. Opening range is meaningful (not too tight, not too wide)
open_price  = prices[:, 0]
eligible    = (
    (open_price >= MIN_PRICE) &
    (or_range_pct >= MIN_RANGE_PCT) &
    (or_range_pct <= MAX_RANGE_PCT)
)

print(f"Total stocks       : {n_stocks}")
print(f"Penny stock filter : {(open_price < MIN_PRICE).sum()} removed")
print(f"Range too tight    : {(or_range_pct < MIN_RANGE_PCT).sum()} removed")
print(f"Range too wide     : {(or_range_pct > MAX_RANGE_PCT).sum()} removed")
print(f"Eligible stocks    : {eligible.sum()}")
print(f"Cost per trade     : ₹{COST_PER_TRADE:,.2f}")
print(f"Hard stop          : {HARD_STOP_PCT*100:.1f}% of entry price")
print(f"Trail gap          : {TRAIL_GAP_PCT*100:.1f}% of entry price")

# ── HELPERS ───────────────────────────────────────────────────────────────
def is_valid_price(p):
    return float(p) > 0 and not np.isnan(float(p))

def calc_raw_pnl(trade_type, entry, curr):
    if not is_valid_price(entry) or not is_valid_price(curr):
        return 0.0
    return ((curr - entry) / entry * position_size if trade_type == "buy"
            else (entry - curr) / entry * position_size)

def close_trade(pos, net_pnl, raw_pnl, curr_price, exit_reason, t):
    global free_units, capital_available, daily_loss, trading_halted

    closed_pnl.append(net_pnl)
    capital_available += unit_size
    free_units += 1
    cooldown[pos["stock"]] = t + COOLDOWN_BARS

    s = pos["stock"]
    if s not in stock_stats:
        stock_stats[s] = {"trades": 0, "pnl": 0.0, "wins": 0}
    stock_stats[s]["trades"] += 1
    stock_stats[s]["pnl"]    += net_pnl
    stock_stats[s]["wins"]   += int(net_pnl > 0)

    if net_pnl < 0:
        daily_loss += abs(net_pnl)
        if daily_loss >= DAILY_LOSS_LIMIT:
            trading_halted = True

    trade_log.append({
        "stock"        : int(s),
        "type"         : pos["type"],
        "entry_time"   : int(pos["entry_time"]),
        "exit_time"    : int(t),
        "bars_held"    : int(t - pos["entry_time"]),
        "entry_price"  : round(float(pos["entry_price"]), 2),
        "exit_price"   : round(float(curr_price), 2),
        "or_high"      : round(float(pos["or_high"]), 2),
        "or_low"       : round(float(pos["or_low"]), 2),
        "peak_pnl"     : round(float(pos["peak_pnl"]), 2),
        "raw_pnl"      : round(float(raw_pnl), 2),
        "costs"        : COST_PER_TRADE,
        "net_pnl"      : round(float(net_pnl), 2),
        "exit_reason"  : exit_reason,
        "breakeven_on" : pos.get("breakeven_active", False),
    })

# ── MAIN LOOP — start after opening range ─────────────────────────────────
for t in range(OPENING_BARS, n_time - 1):

    if trading_halted and len(positions) == 0:
        break

    # ── EXIT ──────────────────────────────────────────────────────────────
    still_open = []

    for pos in positions:
        s          = pos["stock"]
        curr_price = prices[s, t]
        trade_type = pos["type"]
        entry_price= pos["entry_price"]
        bars_held  = t - pos["entry_time"]

        raw_pnl = calc_raw_pnl(trade_type, entry_price, curr_price)
        net_pnl = raw_pnl - COST_PER_TRADE

        # Update peak profit
        pos["peak_pnl"] = max(pos["peak_pnl"], raw_pnl)

        # Price-based stop levels (% of entry price — works across all price ranges)
        hard_stop_price = (
            entry_price * (1 - HARD_STOP_PCT) if trade_type == "buy"
            else entry_price * (1 + HARD_STOP_PCT)
        )

        # Breakeven — once profit hits threshold, floor moves to entry
        profit_pct = raw_pnl / position_size   # as fraction
        if profit_pct >= BREAKEVEN_PCT and not pos.get("breakeven_active"):
            pos["breakeven_active"] = True
            pos["breakeven_price"]  = entry_price  # stop = entry price

        # Trailing stop in price terms
        peak_price = (
            entry_price + (pos["peak_pnl"] / position_size * entry_price)
            if trade_type == "buy"
            else entry_price - (pos["peak_pnl"] / position_size * entry_price)
        )
        trail_stop_price = (
            peak_price * (1 - TRAIL_GAP_PCT) if trade_type == "buy"
            else peak_price * (1 + TRAIL_GAP_PCT)
        )
        trail_active = (pos["peak_pnl"] / position_size) >= TRAIL_ACTIVATE_PCT

        # Effective stop — most favourable of hard stop and breakeven
        if pos.get("breakeven_active"):
            effective_stop = (
                max(hard_stop_price, entry_price) if trade_type == "buy"
                else min(hard_stop_price, entry_price)
            )
        else:
            effective_stop = hard_stop_price

        # Exit conditions
        is_force_eod  = (t >= n_time - 1 - EOD_BUFFER)
        is_hard_stop  = (
            curr_price <= effective_stop if trade_type == "buy"
            else curr_price >= effective_stop
        )
        is_trail_stop = trail_active and (
            curr_price <= trail_stop_price if trade_type == "buy"
            else curr_price >= trail_stop_price
        )

        if is_force_eod:
            close_trade(pos, net_pnl, raw_pnl, curr_price, "force_eod", t)
        elif is_hard_stop:
            close_trade(pos, net_pnl, raw_pnl, curr_price, "hard_stop", t)
        elif is_trail_stop:
            close_trade(pos, net_pnl, raw_pnl, curr_price, "trail_stop", t)
        else:
            still_open.append(pos)

    positions = still_open

    # ── ENTRY ─────────────────────────────────────────────────────────────
    near_eod         = t >= (n_time - 1 - EOD_BUFFER)
    insufficient_cap = capital_available < unit_size

    if trading_halted or free_units == 0 or near_eod or insufficient_cap:
        continue

    open_stocks = {p["stock"] for p in positions}
    open_buys   = sum(1 for p in positions if p["type"] == "buy")
    open_sells  = sum(1 for p in positions if p["type"] == "sell")

    # Rank by breakout magnitude — strongest breakout gets capital first
    breakout_scores = []
    for s in range(n_stocks):
        if not eligible[s]:
            continue
        if s in open_stocks:
            continue
        if cooldown.get(s, 0) > t:
            continue
        if not is_valid_price(prices[s, t]):
            continue

        curr  = prices[s, t]
        score = 0.0
        direction = None

        # Breakout above opening range high → BUY signal
        if curr > or_high[s]:
            score     = (curr - or_high[s]) / or_high[s]   # % above range
            direction = "buy"

        # Breakdown below opening range low → SELL signal
        elif curr < or_low[s]:
            score     = (or_low[s] - curr) / or_low[s]     # % below range
            direction = "sell"

        if direction:
            breakout_scores.append((s, direction, score))

    # Sort by strongest breakout first
    breakout_scores.sort(key=lambda x: x[2], reverse=True)

    for s, direction, score in breakout_scores:
        if free_units == 0 or capital_available < unit_size:
            break

        if direction == "buy":
            if open_buys >= MAX_BUYS:
                continue
            positions.append({
                "stock"            : s,
                "type"             : "buy",
                "entry_price"      : prices[s, t],
                "entry_time"       : t,
                "or_high"          : or_high[s],
                "or_low"           : or_low[s],
                "peak_pnl"         : 0.0,
                "breakeven_active" : False,
            })
            open_stocks.add(s)
            open_buys += 1
            capital_available -= unit_size
            free_units -= 1

        elif direction == "sell":
            if open_sells >= MAX_SELLS:
                continue
            positions.append({
                "stock"            : s,
                "type"             : "sell",
                "entry_price"      : prices[s, t],
                "entry_time"       : t,
                "or_high"          : or_high[s],
                "or_low"           : or_low[s],
                "peak_pnl"         : 0.0,
                "breakeven_active" : False,
            })
            open_stocks.add(s)
            open_sells += 1
            capital_available -= unit_size
            free_units -= 1

# ── RESULTS ───────────────────────────────────────────────────────────────
valid_pnl   = [p for p in closed_pnl if not np.isnan(p)]
total_pnl   = round(np.sum(valid_pnl), 2)
num_trades  = len(trade_log)
avg_pnl     = round(np.mean(valid_pnl), 2) if valid_pnl else 0
wins        = sum(1 for p in valid_pnl if p > 0)
losses      = sum(1 for p in valid_pnl if p <= 0)
total_costs = round(COST_PER_TRADE * num_trades, 2)
best_trade  = round(max(valid_pnl), 2) if valid_pnl else 0
worst_trade = round(min(valid_pnl), 2) if valid_pnl else 0
avg_bars    = round(np.mean([t["bars_held"] for t in trade_log]), 1) if trade_log else 0
be_count    = sum(1 for t in trade_log if t["breakeven_on"])

exit_counts = {}
for tr in trade_log:
    exit_counts[tr["exit_reason"]] = exit_counts.get(tr["exit_reason"], 0) + 1

print("\n" + "=" * 54)
print(f"  Total Net PnL      : ₹{total_pnl:>10,.2f}")
print(f"  Total Costs        : ₹{total_costs:>10,.2f}")
print(f"  Capital Remaining  : ₹{capital_available:>10,.2f}")
print(f"  Daily Loss         : ₹{daily_loss:>10,.2f}  "
      f"{'⚠ HALTED' if trading_halted else '✓ OK'}")
print("-" * 54)
print(f"  Trades             : {num_trades}"
      + (f"  ({wins}W / {losses}L)  WR: {wins/num_trades*100:.1f}%"
         if num_trades else ""))
print(f"  Avg PnL / Trade    : ₹{avg_pnl:>10,.2f}")
print(f"  Best Trade         : ₹{best_trade:>10,.2f}")
print(f"  Worst Trade        : ₹{worst_trade:>10,.2f}")
print(f"  Avg Bars Held      : {avg_bars} bars")
print(f"  Breakeven used     : {be_count} trades")
print("-" * 54)
for reason, count in sorted(exit_counts.items(), key=lambda x: -x[1]):
    pct = count / num_trades * 100 if num_trades else 0
    print(f"  Exit {reason:<16}: {count:>3}  ({pct:.0f}%)")
print("=" * 54)

# Per-stock breakdown
if stock_stats:
    df_stocks = pd.DataFrame([
        {"stock": s, "trades": v["trades"],
         "wins": v["wins"], "net_pnl": round(v["pnl"], 2)}
        for s, v in stock_stats.items()
    ]).sort_values("net_pnl", ascending=False)
    print("\nTop 5 stocks:")
    print(df_stocks.head(5).to_string(index=False))
    print("\nBottom 5 stocks:")
    print(df_stocks.tail(5).to_string(index=False))

print("\nFull Trade Log:")
df = pd.DataFrame(trade_log)
if not df.empty:
    print(df.to_string(index=False))
```

```python
import numpy as np
import pandas as pd

# ── LOAD AND CLEAN ─────────────────────────────────────────────
stock_prices_clean = stock_prices.dropna(axis=0)

# Additional check — ensure no all-zero rows remain
prices_raw = stock_prices_clean.values.astype(float)
valid_rows  = (prices_raw > 0).all(axis=1)
prices      = prices_raw[valid_rows]
n_stocks, n_time = prices.shape

print(f"Stocks after cleaning : {n_stocks}")
print(f"Time bars             : {n_time}")

# ── CONFIG ─────────────────────────────────────────────────────
capital           = 100_000
unit_size         = 20_000          # margin per trade (cash locked)
leverage          = 5
position_size     = unit_size * leverage   # ₹1,00,000 actual exposure

SPIKE_THRESHOLD   = 0.025
SPIKE_STRONG      = 0.050

TRAIL_GAP_PCT     = 0.010
TRAIL_ACTIVATE_PCT= 0.010
BREAKEVEN_PCT     = 0.015
MIN_HOLD_BARS     = 15
TARGET_HOLD_BARS  = 40
MAX_HOLD_BARS     = 150
NO_PROGRESS_BARS  = 20
NO_PROGRESS_MIN   = 0.003

NET_LOSS_LIMIT    = 5_000
EOD_BUFFER        = 40
COOLDOWN_BARS     = 20
MIN_PRICE         = 10.0
MAX_BUYS          = 3
MAX_SELLS         = 3

max_units         = 5
free_units        = max_units
positions         = []
closed_pnl        = []
trade_log         = []
running_net       = 0.0
trading_halted    = False
capital_available = capital
cooldown          = {}
stock_stats       = {}
stock_loss_count  = {}

# ── DYNAMIC HARD STOP ─────────────────────────────────────────
def get_hard_stop_pct(entry_price):
    if entry_price < 20:
        return 0.030
    elif entry_price < 50:
        return 0.025
    elif entry_price < 200:
        return 0.020
    else:
        return 0.015

# ── COST MODEL ────────────────────────────────────────────────
def compute_costs(pos_val):
    buy_val = sell_val = pos_val
    brokerage    = 40
    stt          = 0.00025   * sell_val
    exchange_fee = 0.0000345 * (buy_val + sell_val)
    sebi         = 0.000001  * (buy_val + sell_val)
    stamp        = 0.00003   * buy_val
    gst          = 0.18 * (brokerage + exchange_fee + sebi)
    slippage     = 0.0005    * (buy_val + sell_val)
    return round(brokerage + stt + exchange_fee + sebi + stamp + gst + slippage, 2)

COST_PER_TRADE     = compute_costs(position_size)
MAX_LOSS_PER_TRADE = position_size * 0.030 + COST_PER_TRADE

print(f"Strategy         : Spike fade (counter-trend)")
print(f"Spike threshold  : {SPIKE_THRESHOLD*100:.1f}%  |  Strong: {SPIKE_STRONG*100:.1f}%")
print(f"Hard stop        : dynamic (1.5%-3% by price), no bar delay")
print(f"Trail gap        : {TRAIL_GAP_PCT*100:.1f}% from peak")
print(f"Breakeven        : after {BREAKEVEN_PCT*100:.1f}% profit")
print(f"No-progress exit : after {NO_PROGRESS_BARS} bars <{NO_PROGRESS_MIN*100:.1f}% profit")
print(f"Max hold         : {MAX_HOLD_BARS} bars")
print(f"EOD buffer       : {EOD_BUFFER} bars")
print(f"Stock loss block : after 2 losses on same stock")
print(f"Net loss limit   : ₹{NET_LOSS_LIMIT:,}")
print(f"Cost per trade   : ₹{COST_PER_TRADE:,.2f}")
print(f"Margin per trade : ₹{unit_size:,}  |  Exposure: ₹{position_size:,}")

# ── HELPERS ───────────────────────────────────────────────────
def is_valid(p):
    """Price is valid if positive and not NaN."""
    return float(p) > 0 and not np.isnan(float(p))

def calc_raw_pnl(trade_type, entry, curr):
    if not is_valid(entry) or not is_valid(curr):
        return 0.0
    return ((curr - entry) / entry * position_size if trade_type == "buy"
            else (entry - curr) / entry * position_size)

def calc_profit_pct(trade_type, entry, curr):
    if not is_valid(entry) or entry == 0:
        return 0.0
    return (curr - entry) / entry if trade_type == "buy" else (entry - curr) / entry

def close_trade(pos, net_pnl, rp, curr_price, reason, t):
    global free_units, capital_available, running_net, trading_halted

    closed_pnl.append(net_pnl)
    capital_available += unit_size
    free_units += 1
    cooldown[pos["stock"]] = t + COOLDOWN_BARS

    s = pos["stock"]
    if s not in stock_stats:
        stock_stats[s] = {"trades": 0, "pnl": 0.0, "wins": 0}
    stock_stats[s]["trades"] += 1
    stock_stats[s]["pnl"]    += net_pnl
    stock_stats[s]["wins"]   += int(net_pnl > 0)

    # Track consecutive losses — reset on win
    if net_pnl < 0:
        stock_loss_count[s] = stock_loss_count.get(s, 0) + 1
    else:
        stock_loss_count[s] = 0

    running_net += net_pnl
    if running_net <= -NET_LOSS_LIMIT:
        trading_halted = True

    trade_log.append({
        "stock"        : int(s),
        "type"         : pos["type"],
        "signal"       : pos.get("signal", "normal"),
        "spike_pct"    : round(float(pos.get("spike_pct", 0)), 2),
        "entry_t"      : int(pos["entry_time"]),
        "exit_t"       : int(t),
        "bars_held"    : int(t - pos["entry_time"]),
        "entry_px"     : round(float(pos["entry_price"]), 2),
        "exit_px"      : round(float(curr_price), 2),
        "peak_px"      : round(float(pos["peak_price"]), 2),
        "raw_pnl"      : round(float(rp), 2),
        "costs"        : COST_PER_TRADE,
        "net_pnl"      : round(float(net_pnl), 2),
        "exit_reason"  : reason,
        "breakeven_on" : pos.get("breakeven_active", False),
        "running_net"  : round(running_net, 2),
    })

# ── MAIN LOOP ─────────────────────────────────────────────────
for t in range(1, n_time - 1):

    if trading_halted and len(positions) == 0:
        break

    with np.errstate(divide="ignore", invalid="ignore"):
        prev_valid = (prices[:, t-1] > 0) & (~np.isnan(prices[:, t-1]))
        curr_valid = (prices[:, t]   > 0) & (~np.isnan(prices[:, t]))
        bar_move   = np.where(
            prev_valid & curr_valid,
            (prices[:, t] - prices[:, t-1]) / prices[:, t-1] * 100,
            0.0
        )

    # ── EXIT ──────────────────────────────────────────────────
    still_open = []
    for pos in positions:
        s           = pos["stock"]
        curr_price  = prices[s, t]
        entry_price = pos["entry_price"]
        trade_type  = pos["type"]
        bars_held   = t - pos["entry_time"]

        rp  = calc_raw_pnl(trade_type, entry_price, curr_price)
        net = rp - COST_PER_TRADE

        # Track peak in price terms
        if trade_type == "buy":
            pos["peak_price"] = max(pos["peak_price"], curr_price)
        else:
            pos["peak_price"] = min(pos["peak_price"], curr_price)

        peak_profit_pct = calc_profit_pct(trade_type, entry_price, pos["peak_price"])

        # Breakeven — once peak profit hits threshold
        if peak_profit_pct >= BREAKEVEN_PCT and not pos.get("breakeven_active"):
            pos["breakeven_active"] = True

        # Dynamic hard stop based on entry price level
        stop_pct        = get_hard_stop_pct(entry_price)
        hard_stop_price = (
            entry_price * (1 - stop_pct) if trade_type == "buy"
            else entry_price * (1 + stop_pct)
        )

        # Effective stop — breakeven raises floor to entry price
        if pos.get("breakeven_active"):
            effective_stop = (
                max(hard_stop_price, entry_price) if trade_type == "buy"
                else min(hard_stop_price, entry_price)
            )
        else:
            effective_stop = hard_stop_price

        # Trailing stop from peak price
        active_trail_gap = (
            TRAIL_GAP_PCT * 0.5
            if bars_held >= TARGET_HOLD_BARS
            else TRAIL_GAP_PCT
        )
        trail_price = (
            pos["peak_price"] * (1 - active_trail_gap) if trade_type == "buy"
            else pos["peak_price"] * (1 + active_trail_gap)
        )
        trail_active = peak_profit_pct >= TRAIL_ACTIVATE_PCT

        # Exit conditions
        is_force_eod = (t >= n_time - 1 - EOD_BUFFER)

        # BUG 2 FIX — no bar delay on hard stop
        is_hard_stop = (
            curr_price <= effective_stop if trade_type == "buy"
            else curr_price >= effective_stop
        )

        is_trail_stop = (
            trail_active and
            bars_held >= MIN_HOLD_BARS and
            (curr_price <= trail_price if trade_type == "buy"
             else curr_price >= trail_price)
        )

        is_no_progress = (
            bars_held >= NO_PROGRESS_BARS and
            peak_profit_pct < NO_PROGRESS_MIN
        )

        is_max_hold = bars_held >= MAX_HOLD_BARS

        # Priority: EOD → hard stop → trail → no-progress → max hold
        if is_force_eod:
            close_trade(pos, net, rp, curr_price, "force_eod", t)
        elif is_hard_stop:
            close_trade(pos, net, rp, curr_price, "hard_stop", t)
        elif is_trail_stop:
            close_trade(pos, net, rp, curr_price, "trail_stop", t)
        elif is_no_progress:
            close_trade(pos, net, rp, curr_price, "no_progress", t)
        elif is_max_hold:
            close_trade(pos, net, rp, curr_price, "max_hold", t)
        else:
            still_open.append(pos)

    positions = still_open

    # ── ENTRY ─────────────────────────────────────────────────
    near_eod         = t >= (n_time - 1 - EOD_BUFFER)
    insufficient_cap = capital_available < unit_size

    if trading_halted or free_units == 0 or near_eod or insufficient_cap:
        continue

    if running_net - MAX_LOSS_PER_TRADE <= -NET_LOSS_LIMIT:
        continue

    open_stocks = {p["stock"] for p in positions}
    open_buys   = sum(1 for p in positions if p["type"] == "buy")
    open_sells  = sum(1 for p in positions if p["type"] == "sell")

    candidates = []
    for s in range(n_stocks):
        if s in open_stocks or cooldown.get(s, 0) > t:
            continue

        # BUG 1 FIX — validate current bar price, not opening price
        curr_price_s = prices[s, t]
        if not is_valid(curr_price_s):
            continue
        if curr_price_s < MIN_PRICE:
            continue

        # Block if lost twice on this stock today
        if stock_loss_count.get(s, 0) >= 2:
            continue

        # BUG 5 FIX — only use bar_move if previous price was also valid
        if not is_valid(prices[s, t-1]):
            continue

        move = bar_move[s]

        if move >= SPIKE_STRONG * 100:
            candidates.append((s, "sell", abs(move) + 10, "strong", move))
        elif move >= SPIKE_THRESHOLD * 100:
            candidates.append((s, "sell", abs(move), "normal", move))
        elif move <= -SPIKE_STRONG * 100:
            candidates.append((s, "buy", abs(move) + 10, "strong", move))
        elif move <= -SPIKE_THRESHOLD * 100:
            candidates.append((s, "buy", abs(move), "normal", move))

    candidates.sort(key=lambda x: x[2], reverse=True)

    for s, direction, score, signal, spike_pct in candidates:
        if free_units == 0 or capital_available < unit_size:
            break
        if direction == "buy"  and open_buys  >= MAX_BUYS:
            continue
        if direction == "sell" and open_sells >= MAX_SELLS:
            continue

        entry_px = prices[s, t]
        positions.append({
            "stock"            : s,
            "type"             : direction,
            "signal"           : signal,
            "spike_pct"        : spike_pct,
            "entry_price"      : entry_px,
            "entry_time"       : t,
            "peak_price"       : entry_px,
            "breakeven_active" : False,
        })
        open_stocks.add(s)
        if direction == "buy":
            open_buys += 1
        else:
            open_sells += 1
        capital_available -= unit_size
        free_units -= 1

# ── RESULTS ───────────────────────────────────────────────────
valid_pnl     = [p for p in closed_pnl if not np.isnan(p)]
total_pnl     = round(sum(valid_pnl), 2)
num_trades    = len(trade_log)
avg_pnl       = round(np.mean(valid_pnl), 2) if valid_pnl else 0
wins          = sum(1 for p in valid_pnl if p > 0)
losses        = sum(1 for p in valid_pnl if p <= 0)
total_costs   = round(COST_PER_TRADE * num_trades, 2)
best_trade    = round(max(valid_pnl), 2) if valid_pnl else 0
worst_trade   = round(min(valid_pnl), 2) if valid_pnl else 0
avg_bars      = round(np.mean([t["bars_held"] for t in trade_log]), 1) if trade_log else 0
strong_trades = sum(1 for t in trade_log if t["signal"] == "strong")
be_count      = sum(1 for t in trade_log if t["breakeven_on"])
blocked_stocks= sum(1 for s, c in stock_loss_count.items() if c >= 2)

exit_counts = {}
for tr in trade_log:
    exit_counts[tr["exit_reason"]] = exit_counts.get(tr["exit_reason"], 0) + 1

avg_win  = np.mean([t["net_pnl"] for t in trade_log if t["net_pnl"] > 0]) if wins  else 0
avg_loss = np.mean([t["net_pnl"] for t in trade_log if t["net_pnl"] < 0]) if losses else 0
wl_ratio = abs(avg_win / avg_loss) if avg_loss != 0 else 0

print("\n" + "=" * 58)
print(f"  Total Net PnL      : ₹{total_pnl:>10,.2f}  "
      f"{'PROFIT' if total_pnl > 0 else 'LOSS'}")
print(f"  Total Costs        : ₹{total_costs:>10,.2f}")
print(f"  Margin used/trade  : ₹{unit_size:,}  (exposure ₹{position_size:,})")
print(f"  Capital Remaining  : ₹{capital_available:>10,.2f}")
print(f"  Running Net        : ₹{running_net:>10,.2f}  "
      f"{'⚠ HALTED' if trading_halted else '✓ OK'}")
print("-" * 58)
print(f"  Trades             : {num_trades}"
      + (f"  ({wins}W / {losses}L)  WR: {wins/num_trades*100:.1f}%"
         if num_trades else ""))
print(f"  Strong spikes      : {strong_trades} trades")
print(f"  Stocks blocked     : {blocked_stocks} (2+ losses)")
print(f"  Avg win            : ₹{avg_win:>10,.2f}")
print(f"  Avg loss           : ₹{avg_loss:>10,.2f}")
print(f"  Win / loss ratio   : {wl_ratio:.2f}x")
print(f"  Best trade         : ₹{best_trade:>10,.2f}")
print(f"  Worst trade        : ₹{worst_trade:>10,.2f}")
print(f"  Avg bars held      : {avg_bars} bars")
print(f"  Breakeven used     : {be_count} trades")
print("-" * 58)
for reason, count in sorted(exit_counts.items(), key=lambda x: -x[1]):
    pct = count / num_trades * 100 if num_trades else 0
    print(f"  Exit {reason:<16}: {count:>3}  ({pct:.0f}%)")
print("=" * 58)

if stock_stats:
    df_s = pd.DataFrame([
        {"stock": s, "trades": v["trades"],
         "wins": v["wins"], "net_pnl": round(v["pnl"], 2)}
        for s, v in stock_stats.items()
    ]).sort_values("net_pnl", ascending=False)
    print("\nTop 5 stocks:")
    print(df_s.head(5).to_string(index=False))
    print("\nBottom 5 stocks:")
    print(df_s.tail(5).to_string(index=False))

print("\nFull Trade Log:")
df = pd.DataFrame(trade_log)
if not df.empty:
    print(df.to_string(index=False))
```

```python
import requests
import pandas as pd
import numpy as np
import time
from datetime import datetime, date
import warnings
warnings.filterwarnings("ignore")

# ── UPSTOX CREDENTIALS ────────────────────────────────────────
ACCESS_TOKEN = "eyJ0eXAiOiJKV1QiLCJrZXlfaWQiOiJza192MS4wIiwiYWxnIjoiSFMyNTYifQ.eyJzdWIiOiI0VkFHVVEiLCJqdGkiOiI2OWJlYjQwMDg3YTgwYjEzMWJkZTg3OTgiLCJpc011bHRpQ2xpZW50IjpmYWxzZSwiaXNQbHVzUGxhbiI6dHJ1ZSwiaXNFeHRlbmRlZCI6dHJ1ZSwiaWF0IjoxNzc0MTA1NjAwLCJpc3MiOiJ1ZGFwaS1nYXRld2F5LXNlcnZpY2UiLCJleHAiOjE4MDU2NjY0MDB9.OlQn7q5SG3HnQtLOHS9qnOyvb72CodqfeKadxi0PGxg"   # paste your access token here

HEADERS = {
    "Authorization": f"Bearer {ACCESS_TOKEN}",
    "Accept"       : "application/json",
}

# ── LOAD TICKERS ──────────────────────────────────────────────
tickers = pd.read_csv(
    "/Users/chaitanya/CodeBillion/CodeBillion/dataset/equity_tickers.csv"
)
tickers.columns = tickers.columns.str.strip()   # remove leading/trailing spaces
tickers         = tickers[tickers["SERIES"] == "EQ"].reset_index(drop=True)

print(f"Total EQ stocks : {len(tickers)}")

# ── BUILD INSTRUMENT KEYS ─────────────────────────────────────
# Upstox format: NSE_EQ|ISIN
tickers["instrument_key"] = "NSE_EQ|" + tickers["ISIN NUMBER"].str.strip()

# ── CONFIG ────────────────────────────────────────────────────
CANDLE_INTERVAL  = "1minute"      # matches our backtest
BATCH_SIZE       = 50             # Upstox allows max 50 instruments per request
RATE_LIMIT_SLEEP = 0.5            # seconds between batches — stay under rate limit
TODAY            = date.today().strftime("%Y-%m-%d")

# ── HELPER: FETCH ONE BATCH OF CANDLES ───────────────────────
def fetch_candles_batch(instrument_keys, from_date, to_date, interval):
    """
    Fetch historical candle data for a batch of instruments.
    Returns dict: {instrument_key: list of [timestamp, o, h, l, c, volume]}
    """
    url = "https://api.upstox.com/v2/historical-candle/intraday"

    # Upstox batch endpoint accepts comma-separated instrument keys
    params = {
        "instrument_key" : ",".join(instrument_keys),
        "interval"       : interval,
        "from_date"      : from_date,
        "to_date"        : to_date,
    }

    try:
        response = requests.get(url, headers=HEADERS, params=params, timeout=30)

        if response.status_code == 200:
            data = response.json()
            if data.get("status") == "success":
                return data.get("data", {})
            else:
                print(f"  API error: {data.get('message', 'unknown')}")
                return {}

        elif response.status_code == 429:
            print("  Rate limit hit — sleeping 5 seconds")
            time.sleep(5)
            return fetch_candles_batch(instrument_keys, from_date, to_date, interval)

        else:
            print(f"  HTTP {response.status_code}: {response.text[:200]}")
            return {}

    except requests.exceptions.Timeout:
        print("  Request timed out — skipping batch")
        return {}
    except Exception as e:
        print(f"  Exception: {e}")
        return {}

# ── HELPER: FETCH FULL INSTRUMENT LIST FROM UPSTOX ───────────
def fetch_upstox_instruments():
    """
    Download the full NSE instrument list from Upstox.
    Used to validate instrument keys before fetching.
    """
    url = "https://api.upstox.com/v2/instruments"
    params = {"exchange": "NSE"}

    try:
        response = requests.get(url, headers=HEADERS, params=params, timeout=60)
        if response.status_code == 200:
            instruments = response.json()
            df = pd.DataFrame(instruments)
            print(f"Upstox instruments fetched: {len(df)}")
            return df
        else:
            print(f"Failed to fetch instruments: {response.status_code}")
            return pd.DataFrame()
    except Exception as e:
        print(f"Exception fetching instruments: {e}")
        return pd.DataFrame()

# ── STEP 1: VALIDATE INSTRUMENT KEYS ─────────────────────────
print("\nStep 1 — Fetching Upstox instrument list for validation...")
instruments_df = fetch_upstox_instruments()

if not instruments_df.empty and "isin" in instruments_df.columns:
    valid_isins = set(instruments_df["isin"].dropna().str.strip())
    tickers["valid"] = tickers["ISIN NUMBER"].str.strip().isin(valid_isins)
    valid_tickers    = tickers[tickers["valid"]].reset_index(drop=True)
    invalid_count    = len(tickers) - len(valid_tickers)
    print(f"Valid instruments   : {len(valid_tickers)}")
    print(f"Invalid / not found : {invalid_count}")
else:
    # If instrument fetch fails, proceed with all tickers
    print("Instrument validation skipped — proceeding with all tickers")
    valid_tickers = tickers.copy()

# ── STEP 2: BATCH FETCH CANDLE DATA ───────────────────────────
print(f"\nStep 2 — Fetching {CANDLE_INTERVAL} candles for {TODAY}...")
print(f"Batch size : {BATCH_SIZE} | Total batches : {len(valid_tickers) // BATCH_SIZE + 1}")

all_keys    = valid_tickers["instrument_key"].tolist()
all_symbols = valid_tickers["SYMBOL"].tolist()
all_isins   = valid_tickers["ISIN NUMBER"].str.strip().tolist()

# Storage — dict keyed by instrument_key
raw_candles = {}
failed      = []

total_batches = (len(all_keys) + BATCH_SIZE - 1) // BATCH_SIZE

for batch_num in range(total_batches):
    start = batch_num * BATCH_SIZE
    end   = min(start + BATCH_SIZE, len(all_keys))
    batch_keys = all_keys[start:end]

    print(f"  Batch {batch_num+1:>3}/{total_batches} | "
          f"stocks {start+1}-{end}", end=" ... ")

    result = fetch_candles_batch(
        instrument_keys = batch_keys,
        from_date       = TODAY,
        to_date         = TODAY,
        interval        = CANDLE_INTERVAL,
    )

    if result:
        raw_candles.update(result)
        print(f"got {len(result)} stocks")
    else:
        failed.extend(batch_keys)
        print("failed")

    time.sleep(RATE_LIMIT_SLEEP)

print(f"\nFetch complete:")
print(f"  Successful : {len(raw_candles)} stocks")
print(f"  Failed     : {len(failed)} stocks")

# ── STEP 3: BUILD PRICE MATRIX ────────────────────────────────
print("\nStep 3 — Building price matrix...")

# Find the common set of timestamps across all stocks
# Use closing prices to match our backtest structure
all_timestamps = set()
for key, candles in raw_candles.items():
    for candle in candles:
        all_timestamps.add(candle[0])   # index 0 = timestamp

timestamps_sorted = sorted(all_timestamps)
n_time = len(timestamps_sorted)
ts_index = {ts: i for i, ts in enumerate(timestamps_sorted)}

print(f"Time bars found : {n_time}")
print(f"First bar       : {timestamps_sorted[0] if timestamps_sorted else 'none'}")
print(f"Last bar        : {timestamps_sorted[-1] if timestamps_sorted else 'none'}")

# Build symbol → row mapping
symbol_to_row = {sym: i for i, sym in enumerate(all_symbols)}
isin_to_symbol = dict(zip(all_isins, all_symbols))

# Instrument key → symbol mapping
key_to_symbol = {}
for _, row in valid_tickers.iterrows():
    key_to_symbol[row["instrument_key"]] = row["SYMBOL"]

# Initialise price matrix with NaN
n_stocks_fetched = len(all_symbols)
price_matrix     = np.full((n_stocks_fetched, n_time), np.nan)

# Fill in closing prices
filled_count = 0
for key, candles in raw_candles.items():
    symbol = key_to_symbol.get(key)
    if symbol is None:
        continue
    row_idx = symbol_to_row.get(symbol)
    if row_idx is None:
        continue

    for candle in candles:
        ts      = candle[0]   # timestamp
        close   = candle[4]   # index 4 = close price
        col_idx = ts_index.get(ts)
        if col_idx is not None:
            price_matrix[row_idx, col_idx] = close
            filled_count += 1

print(f"Prices filled   : {filled_count:,}")
print(f"NaN count       : {np.isnan(price_matrix).sum():,}")

# ── STEP 4: BUILD DATAFRAME ───────────────────────────────────
stock_prices_live = pd.DataFrame(
    price_matrix,
    index   = all_symbols,
    columns = [str(i) for i in range(n_time)]
)

print(f"\nFinal price matrix shape : {stock_prices_live.shape}")
print(f"Stocks with full data    : {stock_prices_live.dropna(axis=0).shape[0]}")
print(f"Stocks with partial data : {stock_prices_live.isna().any(axis=1).sum()}")

# ── STEP 5: SAVE AND PREVIEW ──────────────────────────────────
save_path = f"/Users/chaitanya/CodeBillion/CodeBillion/dataset/live_{TODAY}.csv"
stock_prices_live.to_csv(save_path)
print(f"\nSaved to : {save_path}")

print("\nSample (first 3 stocks, first 5 bars):")
print(stock_prices_live.iloc[:3, :5])
#
print("\nReady to use with strategy — assign:")
print("  stock_prices = stock_prices_live")
```
