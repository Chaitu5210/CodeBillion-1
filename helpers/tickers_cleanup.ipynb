```python
from tqdm import tqdm
import pandas as pd
import requests
import time

equity_tickers = pd.read_csv("../dataset/equity_tickers.csv")
equity_tickers["INSTRUMENT_KEY"] = "NSE_EQ|" + equity_tickers[" ISIN NUMBER"]

date = "2025-03-20"
access_token = ""

headers = {
    "Authorization": f"Bearer {access_token}"
}

base_url = "https://api.upstox.com/v2/historical-candle"

all_prices = []

# 👇 tqdm added here
for instrument in tqdm(equity_tickers["INSTRUMENT_KEY"], desc="Fetching data"):

    try:
        url = f"{base_url}/{instrument}/1minute/{date}/{date}"
        response = requests.get(url, headers=headers)

        data = response.json()

        if "data" not in data or "candles" not in data["data"]:
            all_prices.append([])
            continue

        candles = data["data"]["candles"]
        close_prices = [c[4] for c in candles]

        all_prices.append(close_prices)

        time.sleep(0.2)

    except Exception as e:
        print(f"Error with {instrument}: {e}")
        all_prices.append([])
```
